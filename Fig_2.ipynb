```python
!apt-get install -y fonts-liberation 2>/dev/null
import matplotlib
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf')
fm.fontManager.addfont('/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf')
matplotlib.rcParams['font.family'] = 'Liberation Sans'
matplotlib.rcParams['font.family'] = 'Liberation Sans'
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import rcParams

# PLOS-compliant font choice (use Arial if available on your system)
rcParams["font.size"] = 12

fig, ax = plt.subplots(figsize=(12, 10), dpi=300)
ax.set_xlim(0, 12)
ax.set_ylim(0, 10)
ax.axis("off")

# Colors
layer1_color = "#E8F4FD"
layer2_color = "#FFF3E0"
layer3_color = "#E8F5E9"
arrow_color = "#333333"
box_color = "#FFFFFF"
border_color = "#666666"
note_color = "#777777"

# -------------------------
# Layer 1: Data Ingestion
# -------------------------
layer1 = mpatches.FancyBboxPatch(
    (0.5, 7.0), 11, 2.5,
    boxstyle="round,pad=0.2",
    facecolor=layer1_color,
    edgecolor=border_color,
    linewidth=2
)
ax.add_patch(layer1)
ax.text(
    6, 9.15, "Layer 1: Data Ingestion",
    fontsize=12, fontweight="bold",
    ha="center", va="center"
)

layer1_boxes = [
    (1.5, "NOAA Data"),
    (4.5, "QC & Conversion"),
    (7.5, "Time Alignment"),
    (10.0, "Hashing"),
]

for x, label in layer1_boxes:
    box = mpatches.FancyBboxPatch(
        (x - 0.9, 7.35), 1.8, 1.3,
        boxstyle="round,pad=0.1",
        facecolor=box_color,
        edgecolor=border_color,
        linewidth=1
    )
    ax.add_patch(box)
    ax.text(x, 8.0, label, fontsize=12, ha="center", va="center")

for x_start in [2.4, 5.4, 8.4]:
    ax.annotate(
        "", xy=(x_start + 0.7, 8.0), xytext=(x_start, 8.0),
        arrowprops=dict(arrowstyle="->", color=arrow_color, lw=1.5)
    )

ax.annotate(
    "", xy=(6, 6.8), xytext=(6, 7.0),
    arrowprops=dict(arrowstyle="->", color=arrow_color, lw=2)
)

# -------------------------
# Layer 2: Blockchain Verification
# -------------------------
layer2 = mpatches.FancyBboxPatch(
    (0.5, 4.0), 11, 2.8,
    boxstyle="round,pad=0.2",
    facecolor=layer2_color,
    edgecolor=border_color,
    linewidth=2
)
ax.add_patch(layer2)
ax.text(
    6, 6.45, "Layer 2: Blockchain Verification",
    fontsize=12, fontweight="bold",
    ha="center", va="center"
)

layer2_boxes = [
    (2.5, "On-Chain Hashes"),
    (6.0, "PoS Validation"),
    (9.5, "IPFS Records"),
]

for x, label in layer2_boxes:
    box = mpatches.FancyBboxPatch(
        (x - 1.2, 4.35), 2.4, 1.5,
        boxstyle="round,pad=0.1",
        facecolor=box_color,
        edgecolor=border_color,
        linewidth=1
    )
    ax.add_patch(box)
    ax.text(x, 5.1, label, fontsize=12, ha="center", va="center")

ax.annotate(
    "", xy=(4.8, 5.1), xytext=(3.7, 5.1),
    arrowprops=dict(arrowstyle="<->", color=arrow_color, lw=1.5)
)
ax.annotate(
    "", xy=(8.3, 5.1), xytext=(7.2, 5.1),
    arrowprops=dict(arrowstyle="<->", color=arrow_color, lw=1.5)
)

ax.annotate(
    "", xy=(6, 3.8), xytext=(6, 4.0),
    arrowprops=dict(arrowstyle="->", color=arrow_color, lw=2)
)

# -------------------------
# Layer 3: Statistical Analysis
# -------------------------
layer3 = mpatches.FancyBboxPatch(
    (0.5, 1.0), 11, 2.8,
    boxstyle="round,pad=0.2",
    facecolor=layer3_color,
    edgecolor=border_color,
    linewidth=2
)
ax.add_patch(layer3)
ax.text(
    6, 3.45, "Layer 3: Statistical Analysis",
    fontsize=12, fontweight="bold",
    ha="center", va="center"
)

layer3_boxes = [
    (2.5, "Z-Score"),
    (6.0, "Anomaly Rule"),
    (9.5, "Flagged Records"),
]

for x, label in layer3_boxes:
    box = mpatches.FancyBboxPatch(
        (x - 1.2, 1.35), 2.4, 1.5,
        boxstyle="round,pad=0.1",
        facecolor=box_color,
        edgecolor=border_color,
        linewidth=1
    )
    ax.add_patch(box)
    ax.text(x, 2.1, label, fontsize=12, ha="center", va="center")

ax.annotate(
    "", xy=(4.8, 2.1), xytext=(3.7, 2.1),
    arrowprops=dict(arrowstyle="->", color=arrow_color, lw=1.5)
)
ax.annotate(
    "", xy=(8.3, 2.1), xytext=(7.2, 2.1),
    arrowprops=dict(arrowstyle="->", color=arrow_color, lw=1.5)
)

# -------------------------
# Side annotation
# -------------------------
ax.annotate(
    "Amendment\nTrail",
    xy=(11.2, 5.1), xytext=(11.2, 2.1),
    fontsize=9, ha="center", va="bottom",
    fontstyle="italic", color="black",
    arrowprops=dict(
        arrowstyle="->",
        color="black",
        lw=1.2,
        linestyle="dashed"
    )
)

plt.tight_layout()
plt.savefig(
    "Fig2_architecture.tif",
    format="tiff",
    dpi=300,
    pil_kwargs={"compression": "tiff_lzw"}
)
plt.savefig("Fig2_architecture.png", format="png", dpi=300)
plt.show()
```
