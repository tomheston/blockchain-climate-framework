```python
import pandas as pd
import numpy as np
from datetime import datetime

# Parse the NOAA data file
data = []

with open('noaa_984290_2024.txt', 'r') as f:
    for line in f:
        parts = line.split()
        if len(parts) < 5:
            continue

        year = int(parts[0])
        month = int(parts[1])
        day = int(parts[2])
        hour = int(parts[3])

        # Temperature in tenths of degrees C
        temp_raw = int(parts[4])
        if temp_raw == -9999:  # Missing data
            continue

        temp_c = temp_raw / 10.0
        dt = datetime(year, month, day, hour)

        data.append({
            'datetime': dt,
            'temp_c': temp_c
        })

df = pd.DataFrame(data)

# Count total lines and -9999 flags in full file
total_lines = 0
flagged = 0
with open('noaa_984290_2024.txt', 'r') as f:
    for line in f:
        parts = line.split()
        if len(parts) < 5:
            continue
        total_lines += 1
        if int(parts[4]) == -9999:
            flagged += 1
print(f"Total records: {total_lines}")
print(f"Flagged -9999: {flagged}")
print(f"Valid: {total_lines - flagged}")
print(f"Expected hours (2024 leap year): {366*24}")
print(f"Completeness: {total_lines/(366*24)*100:.1f}%")

# Calculate statistics
print(f"\nTotal valid records: {len(df)}")
print(f"Date range: {df['datetime'].min()} to {df['datetime'].max()}")
print(f"Temp range: {df['temp_c'].min():.1f}°C to {df['temp_c'].max():.1f}°C")

mean_temp = df['temp_c'].mean()
std_temp = df['temp_c'].std()

print(f"\nMean: {mean_temp:.2f}°C")
print(f"Std Dev: {std_temp:.2f}°C")

# Calculate z-scores
df['z_score'] = (df['temp_c'] - mean_temp) / std_temp
df['anomaly'] = abs(df['z_score']) > 3.0

n_anomalies = df['anomaly'].sum()
print(f"\nAnomalies detected (|z| > 3.0): {n_anomalies}")

if n_anomalies > 0:
    print("\n" + "="*70)
    print("DETECTED ANOMALIES")
    print("="*70)
    anomalies = df[df['anomaly']].sort_values('z_score', key=abs, ascending=False)

    for idx, row in anomalies.iterrows():
        print(f"\n{row['datetime']}")
        print(f"  Temperature: {row['temp_c']:.1f}°C")
        print(f"  Z-score: {row['z_score']:+.2f}σ")
        print(f"  Deviation: {row['temp_c'] - mean_temp:+.1f}°C from mean")

# Export
df[df['anomaly']].to_csv('anomalies.csv', index=False)
print(f"\n\nResults exported to anomalies.csv")

# Sensitivity analysis
print("\n" + "="*70)
print("SENSITIVITY ANALYSIS")
print("="*70)
for threshold in [2.5, 3.0, 3.5]:
    flagged_obs = df[abs(df['z_score']) > threshold]
    n = len(flagged_obs)
    if n > 0:
        apr = (flagged_obs['datetime'].dt.month == 4).sum()
        may = (flagged_obs['datetime'].dt.month == 5).sum()
        other = n - apr - may
        tmin = flagged_obs['temp_c'].min()
        tmax = flagged_obs['temp_c'].max()
        print(f"\n|z| > {threshold}: {n} anomalies")
        print(f"  Temp range: {tmin:.1f}–{tmax:.1f}°C")
        print(f"  April: {apr}, May: {may}, Other months: {other}")
    else:
        print(f"\n|z| > {threshold}: 0 anomalies")
```
